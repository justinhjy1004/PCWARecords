```python
import re

# Regex pattern for matching the first address
address_pattern = r'^\d+\s[A-Za-z]+\s[A-Za-z\s]+,\s[A-Za-z\s]+,\s(?:WA|Washington)\s\d{5}'

# Find the first match of the address pattern
match = re.search(address_pattern, address, re.MULTILINE | re.IGNORECASE)

if match:
    print("Address found:", match.group())
else:
    pass
```

```python
matches = address_pattern.finditer(address)
```

```python
for match in matches:
    print("Full address:", match.group(0))
    print("Street Number:", match.group('number'))
    print("Street Name:", match.group('street').strip())
    print("Street Type:", match.group('type'))
    print("Direction:", match.group('direction'))
    print("City:", match.group('city').strip())
    print("State:", match.group('state'))
    print("ZIP Code:", match.group('zip'))
    print("-" * 40)
```
